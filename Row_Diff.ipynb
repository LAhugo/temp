```python
def u_f_4(df): # price 1 to 5 f only need price_dif index 2:5, all of 4(them) must <0.06
    return ( 
            # 3 of 4
            ((df.loc[6:9, 'price_dif'] < 0).sum() >= 3       # total 4, need 3, look at the future row for differnce
             and (df.loc[6:9, 'price_dif2'] < 0).sum() >= 1  # make sure continues
             and (df.loc[2:5, 'price_ratio'] <= 0.06).all()   # on the top doesn't count
             and (df.loc[1:4,'price'] < df['price'][0]).all() )  
            
            or 
            ((df.loc[6:10, 'price_dif'] < 0).sum() >= 4             # 4 of 5     
             and (df.loc[6:10, 'price_dif2'] < 0).sum() >= 2  
             and (df.loc[2:5, 'price_ratio'] <= 0.06).all()  
             and (df.loc[1:4,'price'] < df['price'][0]).all() )  
            
            or 
            ((df.loc[6:11, 'price_dif'] < 0).sum() >= 5            #5 of 6
             and (df.loc[6:11, 'price_dif2'] < 0).sum() >= 3  
             and (df.loc[2:5, 'price_ratio'] <= 0.06).all()  
             and (df.loc[1:4,'price'] < df['price'][0]).all() ) 
            
            or 
            ((df.loc[6:12, 'price_dif'] < 0).sum() >= 5            #5 of 7
             and (df.loc[6:12, 'price_dif2'] < 0).sum() >= 3  
             and (df.loc[2:5, 'price_ratio'] <= 0.06).all()  
             and (df.loc[1:4,'price'] < df['price'][0]).all() ) 
            
            or
            ((df.loc[6:13, 'price_dif'] < 0).sum() >= 6
             and (df.loc[6:13, 'price_dif2'] < 0).sum() >= 4  
             and (df.loc[2:5, 'price_ratio'] <= 0.06).all()  
             and (df.loc[1:4,'price'] < df['price'][0]).all() )  
            
            or ((df.loc[6:14, 'price_dif'] < 0).sum() >= 7           
             and (df.loc[6:14, 'price_dif2'] < 0).sum() >= 4  
             and (df.loc[2:5, 'price_ratio'] <= 0.06).all()  
             and (df.loc[1:4,'price'] < df['price'][0]).all() ) 
            
            or ((df.loc[6:15, 'price_dif'] < 0).sum() >= 7         
             and (df.loc[6:15, 'price_dif2'] < 0).sum() >= 4  
             and (df.loc[2:5, 'price_ratio'] <= 0.06).all()  
             and (df.loc[1:4,'price'] < df['price'][0]).all() ) 
            )
```
